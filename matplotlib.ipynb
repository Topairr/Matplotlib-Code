```python
plt.figure(
    figsize=(15, 10)
)
plt.plot(df['Date'], df['steps'], 
         marker='o', 
         linestyle='-',
         color = 'b',
         label='Daily Steps',   
         
         )
plt.title('Daily Steps this December')
plt.xlabel('Date')
plt.ylabel('Number of Steps')
plt.legend() # To dispaly the label attached in the plt.plot() function
plt.show()
```

#### Creating a Bar plot

```python
plt.figure(
    figsize=(17, 10)   
)
plt.bar(df['Date'], df['Ditance Covered (M)'],
        color='g',
        label='Distance Covered (M)',
        
        )
plt.title('Daily Distance Covered this December')
plt.xlabel('Date')
plt.ylabel('Distance Covered (M)')
plt.legend() # To dispaly the label attached in the plt.plot() function
plt.show()
```
